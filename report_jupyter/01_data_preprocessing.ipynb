```python
from utils import load_data, save_data
import utils.data_preprocessing as process
```

# 01 - Data Preprocessing

## 01.01 - Load the Data

```python
filename = 'arxiv'
df = load_data(f'../data/{filename}.csv')
```

## 01.02 - Clean Categories

```python
df = process.categories(df)
df.head()
```

## 01.03 - Remove Duplicates

```python
df = process.duplicates(df)
df.head()
```

## 01.04 - Drop Null or NA Values

```python
df = process.null(df)
df.head()
```

## 01.05 - Lowercase the Text Data

```python
df = process.lowercase(df)
df.head()
```

## 01.06 - Remove Punctuation

```python
df = process.punctuation(df)
df.head()
```

## 01.07 - Remove Numbers

```python
df = process.numbers(df)
df.head()
```

## 01.08 - Remove Extra Whitespace

```python
df = process.whitespace(df)
df.head()
```

## 01.09 - Remove Stopwords

```python
df = process.stopwords(df)
df.head()
```

## 01.10 - Handling Contractions

```python
df = process.contractions(df)
df.head()
```

## 01.11 - Process Diacritics

```python
df = process.diacritics(df)
df.head()
```

## 01.12 - Remove Special Characters

```python
df = process.special_characters(df)
df.head()
```

## 01.13 - Handling Emojis and Emoticons

```python
df = process.emojis(df)
df.head()
```

## 01.14 - Remove HTML Tags

```python
df = process.html(df)
df.head()
```

## 01.15 - Enforce UTF-8 Encoding

```python
df = process.utf8(df)
df.head()
```

## 01.16 - Save the Processed Data

```python
save_data(df, f'../data/01_{filename}_processed.bin')
```
